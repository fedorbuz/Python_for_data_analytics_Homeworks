```python
from bs4 import BeautifulSoup
import requests
import sqlite3
from sqlite3 import Error
import pandas as pd
```

#### Загружаем тестовый файл html - 250 лучших фильмов

```python
url = 'http://www.imdb.com/chart/top'
response = requests.get(url)
soup = BeautifulSoup(response.text, "lxml")
```

#### Парсим интересующие параметры со страницы со списком 250 лучших фильмов

```python
crew = [a.attrs.get('title') for a in soup.select('td.titleColumn a')]
rating = [b.attrs.get('data-value')
        for b in soup.select('td.posterColumn span[name=ir]')]
link = [c.attrs.get('href')
        for c in soup.select('td.posterColumn a')]
date = soup.select('span.secondaryInfo')
movies = [d.attrs.get('alt') for d in soup.select('td.posterColumn img')]
```

```python
movies
```

```python
crew
```

```python
movie_rating = [round(float(r),2) for r in rating]
movie_rating
```

```python
movie_link = ['https://www.imdb.com/' + l for l in link]
movie_link
```

Жанры не указаны на основной странице, достаем их со страницы каждого фильма

```python
genres = []
for i in range(250):
    movie_url = movie_link[i]
    movie_response = requests.get(movie_url)
    movie_soup = BeautifulSoup(movie_response.text, "lxml")
    genre = movie_soup.find('div', class_='ipc-chip-list sc-16ede01-4 bMBIRz')
    if genre is not None:
        movie_genre = genre.text.strip()
        genres.append(movie_genre)
    else:
        genres.append('Unknown')
```

```python
genres
```

```python
release_year = [int(y[1:-1]) for y in [dt.text.strip() for dt in date]]
release_year
```

#### Создаем список из словарей по каждому фильму

```python
movies_list = []
for i in range(len(movies)):
    movie_data = {'title':movies[i],
                'year':release_year[i],
                  'genre':genres[i],
                 'crew':crew[i],
                 'rating':movie_rating[i],
                 'link':movie_link[i]}
    movies_list.append(movie_data)
```

```python
movies_list
```

Выведем список фильмов в табличном виде

```python
movie_df = pd.DataFrame(movies_list)
movie_df
```

#### Создадим БД фильмов 

```python
from sqlalchemy import create_engine
engine = create_engine('sqlite://', echo=False)
```

```python
movie_df.to_sql('movies', con=engine, index=False)
```
